### Synopsis 데이터 토큰화

```python
okt = Okt()
reg = re.compile("[^ㄱ-ㅎㅏ-ㅣ가-힣\s]")
tokenized_synopsis = []
with open("C:/datasets_temp/synopsis.txt", 'r', encoding='utf-8') as file:
    while True:
        line = file.readline()
        if not line: break
        
        # 정규식
        line = reg.sub('', line)
        
        # 형태소 분석
        words = okt.pos(line, stem=True, norm=True)
        res = []
        for w in words:
            if w[1] in ["Noun"]:      # "Adjective", "Verb" 등을 포함할 수도 있습니다.
                res.append(w[0])    # 명사일 때만 tokenized 에 저장하게 됩니다. 
        tokenized_synopsis.append(res)
```

### 데이터 불러오기(23개 클래스의 영화 텍스트)

```python
directory = Path('C:/datasets_temp/synopsis' )
files = [file.name for file in directory.glob("*.txt")]

reg = re.compile("[^ㄱ-ㅎㅏ-ㅣ가-힣\s]")

files_data = {}

for file in files:
    with open(directory / file, "r", encoding="utf-8") as f:
        text = f.read()
        text = reg.sub('', text)
        files_data[file] = text
```

### 23개 클래스 토큰화하기

```python
tokenized_class = {}

for file, text in files_data.items():
    words = okt.pos(text, stem=True, norm=True)
    
    tokens = [w[0] for w in words if w[1] in ['Noun']]
    
    tokenized_class[file] = tokens
```

```python
# TF-IDF DataFrame 변환
feature_names = vectorizer.get_feature_names_out()
df_tfidf = pd.DataFrame(tfidf.toarray(), columns=feature_names, index=tokenized_class.keys())

# synopsis_gen.txt와 synopsis_art.txt의 단어 선정 (서로 겹치지 않도록)
gen_sorted = df_tfidf.loc["synopsis_gen.txt"].sort_values(ascending=False).index.tolist()
art_sorted = df_tfidf.loc["synopsis_art.txt"].sort_values(ascending=False).index.tolist()

# 중복 없는 방식으로 10개씩 선택
gen_words = []
art_words = []
selected_genart_words = set()  # gen-art 간 중복 방지용

for word in gen_sorted:
    if word not in selected_genart_words:
        gen_words.append(word)
        selected_genart_words.add(word)
    if len(gen_words) == 50:
        break

for word in art_sorted:
    if word not in selected_genart_words:
        art_words.append(word)
        selected_genart_words.add(word)
    if len(art_words) == 50:
        break

class_keywords["synopsis_art.txt"] = art_words
class_keywords["synopsis_gen.txt"] = gen_words


# 나머지 장르에 대해 기존 방식 적용 (하지만 gen-art와는 중복 가능)
selected_words = set()  # 나머지 장르들 간의 중복 방지용

for class_name, row in df_tfidf.iterrows():
    if class_name in ["synopsis_art.txt", "synopsis_gen.txt"]:
        continue  # 이미 처리했으므로 스킵
    
    sorted_words = row.sort_values(ascending=False).index.tolist()

    # gen-art에서 선택된 단어는 중복 가능하지만, 나머지 장르끼리는 중복 피해야 함
    unique_words = [word for word in sorted_words if word not in selected_words][:10]
    class_keywords[class_name] = unique_words
    selected_words.update(unique_words)  # 선택된 단어를 전역 집합에 추가
```

#### WEAT-score 계산 및 시각화에 필요한 도구 정의

```python
def cos_sim(i, j):
    return np.dot(i, j.T)/(np.linalg.norm(i)*np.linalg.norm(j))
```

```python
def s(w, A, B):
    c_a = cos_sim(w, A)
    c_b = cos_sim(w, B)
    mean_A = np.mean(c_a, axis=-1)
    mean_B = np.mean(c_b, axis=-1)
    return mean_A - mean_B
```

```python
def weat_score(X, Y, A, B):
    
    s_X = s(X, A, B)
    s_Y = s(Y, A, B)

    mean_X = np.mean(s_X)
    mean_Y = np.mean(s_Y)
    
    std_dev = np.std(np.concatenate([s_X, s_Y], axis=0))
    
    return  (mean_X-mean_Y)/std_dev
```

```python
def get_vectors(words):
    return np.array([model.wv[word] for word in words if word in model.wv])
```

```python
# 히트맵 시각화
def plot_heatmap(matrix_df, title):
    plt.figure(figsize=(10, 8))
    sns.heatmap(matrix_df, annot=False, cmap="coolwarm", fmt=".3f", linewidths=0.5)
    plt.title(title)
    plt.xlabel("Class")
    plt.ylabel("Class")
    plt.xticks(rotation=45)
    plt.yticks(rotation=45)
    plt.show()
```

### 예술 / 일반 영화 편향 찾아보기

```python
# WEAT Score 행렬을 (21, 21) 크기로 설정
n_classes = len(class_keywords.keys()) - 2  # synopsis_gen, synopsis_art 제외
genart_weat_matrix = np.zeros((n_classes, n_classes))

# 비교 대상 장르 리스트 (Gen, Art 제외)
filtered_classes = [c for c in class_keywords.keys() if c not in ["synopsis_gen.txt", "synopsis_art.txt"]]

vectors_gen = get_vectors(class_keywords["synopsis_gen.txt"])
vectors_art = get_vectors(class_keywords["synopsis_art.txt"])

# 각 장르 쌍 (P, Q)에 대해 WEAT Score 계산
for i in range(n_classes):
    for j in range(n_classes):
        if i == j:
            genart_weat_matrix[i, j] = np.nan  # 자기 자신과 비교는 의미 X
        
        class_P = filtered_classes[i]
        class_Q = filtered_classes[j]

        vectors_P = get_vectors(class_keywords[class_P])
        vectors_Q = get_vectors(class_keywords[class_Q])

        # WEAT Score 계산 (P vs Q - General vs Art)
        genart_weat_matrix[i, j] = weat_score(vectors_P, vectors_Q, vectors_gen, vectors_art)

genart_df = pd.DataFrame(
    genart_weat_matrix, 
    index=filtered_classes, 
    columns=filtered_classes
)
```
